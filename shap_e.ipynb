```python
from IPython.display import Image, display
from diffusers import DiffusionPipeline
from diffusers.utils import export_to_gif
import torch
import numpy as np
```

```python
pipe = DiffusionPipeline.from_pretrained(
    "openai/shap-e",
    torch_dtype=torch.float16,
)
pipe.to("cuda")
```

```python
def generate_and_show_images(prompt: str, num_variations: int = 1, dpi: int = 150):
    volumes = pipe(
        prompt=[prompt] * num_variations,
        guidance_scale=5,
        num_inference_steps=25,
        frame_size=256,
    ).images
    for volume in volumes:
        export_to_gif(volume, "volume.gif")
        display(Image(url="volume.gif"))
```

```python
generate_and_show_images("a red apple")
```
